```python
import matplotlib.pyplot as plt
import json
import seaborn as sns
```

```python
data = json.loads(
    open("0.1-moesen-visualize-training.json", "r").read()
)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle("Epochs: 5")
ax1.plot(data["train_l"])
ax1.plot(data["test_l"])
ax1.set_title("Loss")
ax1.legend(["Train", "Test"])

ax2.plot(data["acc"])
ax2.set_title("Accuracy")
ax2.legend(["Test"])

plt.show()
```
